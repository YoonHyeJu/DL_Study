```
#특정 범위 픽셀 변경
import matplotlib.pyplot as plt
import time

image = cv2.imread('unnamed.png')

#방법1
s_time = time.time()
for i in range(100):
    for j in range(100):
        image[i,j] = [255,255,255]
print("------------%s seconds-----------" %(time.time() - s_time))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```
#방법2
s_time = time.time()
image[0:100,0:100] = [150,150,150]
print("------------%s seconds-----------" %(time.time() - s_time))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```
#ROI 추출 및 복사(관심 있는 부분 추출)
image = cv2.imread('unnamed.png')

roi = image[100:350, 100:350]

image[0:250, 0:250] = roi

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```
#픽셀별로 색상 다루기
image = cv2.imread('unnamed.png')

image[:,:,1] = 0 #'G' 값 0으로 만들기

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
