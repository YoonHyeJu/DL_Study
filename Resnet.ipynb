```
from keras.applications import ResNet50
import tensorflow as tf
```

```

##### 모델 구조를 json 파일로 저장하는 코드 #####

resnet = ResNet50(weights = 'imagenet')
resnet.summary()

json_string = resnet.to_json()

with open("model.json", "w") as f:
    f.write(json_string)
```

```
from keras.layers import ZeroPadding2D, Conv2D, BatchNormalization, MaxPooling2D, Dense, Flatten
from keras.models import Model
```

```
model = tf.keras.Sequential()
```

```
input_shape = (224,224,3)

input_ = tf.keras.Input(shape = input_shape)

X = input_

#x = ZeroPadding2D(padding=(2,2))(X)
x = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(X)
x = BatchNormalization()(x)
#x = ZeroPadding2D(padding=(2,2))(x)
x = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(X)
x = MaxPooling2D()(x)
X = x

#x = MaxPooling2D()(X)
x = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(X)
x = BatchNormalization()(x)
x = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(x)
x = BatchNormalization()(x)
x = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(x)
x = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(x)

X = X + x

#x = MaxPooling2D()(X)
x = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(X)
x = Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')(x)
X = x + x

#x = MaxPooling2D()(X)

x = Flatten()(X)
x = Dense(128)(x)
x = Dense(128)(x)
x = Dense(2, activation='sigmoid')(x)

output_ = x
```

```
model = Model(input_, output_)
```
